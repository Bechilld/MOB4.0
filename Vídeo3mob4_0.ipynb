Vamos destacar as colunas, saber quais informações estão em cada coluna, usando "columns".

```python
#mostrando apenas uma coluna específica. 
df.columns
```

Uma pergunta, é possível trabalhar com apenas uma coluna? 
A resposta é sim. E é muito simples:

```python
#mostrando a coluna que contém os preços. coluna price.
df["price"]
```

Como está no código, no indice da variável df é colocado o nome da coluna buscada.  

Uma outra pergunta que pode ser feita, é possível adicionar uma coluna nova ao DataFrame? Novamente a resposta é sim.
Da mesma forma, usando o indice da váriavel:

```python
#Adicionando uma nova coluna.
df["Coluna nova"]=0.0
```

Vamos ver a coluna nova com o que já sabemos.

```python
df["Coluna nova"]
```

Por fim, uma função muito utilizada é quando queremos encontrar o maior valor, o menor valor, o valor médio e etc. 
Exemplo:

```python
#Encontrando o valor médio da coluna dos preços.
df["price"].mean()
```

Usando "mean" deu o valor médio. Mas caso queira um acumulado de informações de imediato, masta usar "describe":


```python
#Usando describe para preço.
df["price"].describe()
```

Num comando só, já aponta o valor máximo, mínimo, desvio padrão(std) entre outros.
